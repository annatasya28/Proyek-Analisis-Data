## Visualization & Explanatory Analysis

### Pertanyaan 1: Pada musim apa penyewaan sepeda paling ramai?

```python
#create bar chart for relationship between season and total rental bikes
season = ('Spring', 'Summer', 'Fall', 'Winter')
total_rental_bikes_season = (471348, 918589, 1061129, 841613)
plt.bar(x=season, height=total_rental_bikes)
plt.xlabel('Season')
plt.ylabel('Total Rental Bikes')
plt.title('Total Rental Bikes per Season')
plt.show()
```

### Pertanyaan 2: Pada cuaca apa penyewaam sepeda paling ramai?

```python
#Create bar chart for relationship between weather and total rental bikes
weather = ('Clear', 'Mist', 'Light Snow', 'Heavy Rain')
total_rental_bikes_weather = (2257952, 996858, 37689, 0)
plt.bar(x=weather, height=total_rental_bikes_weather)
plt.xlabel('Weather')
plt.ylabel('Total Rental Bikes')
plt.title('Total Rental Bikes per Weather')
plt.show()
```

**Insight:**
- Berdasarkan bar chart terlihat jelas bahwa musim dan cuaca mempengaruhi keramaian jumlah penyewa sepeda
- Jumlah penyewa sepeda tertinggi saat musim gugur dan cuaca cerah
- Jumlah penyewa sepeda terendah pada saat musim salju dan cuaca hujan deras
